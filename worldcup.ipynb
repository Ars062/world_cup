```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,axs=plt.subplots(3,2, figsize=(12,10))
fig.suptitle("Distribution",fontsize=20)
data2=np.random.normal(loc=0,scale=0,size=1000)
data3=np.random.uniform(low=0,high=1,size=1000)
data1=np.concatenate([np.random.normal(loc=4,scale=0,size=1000),
                     np.random.normal(loc=-4,scale=0,size=1000)])
data=np.concatenate([np.random.normal(loc=4,scale=1.5,size=1000),
                     np.random.normal(loc=0,scale=0,size=1000),
                     np.random.normal(loc=-4,scale=1.5,size=1000)])
sns.histplot(data3,ax=axs[0,1],color="black",bins=30,kde=True,stat="density")
sns.histplot(data2,ax=axs[0,0],color="brown",bins=30,kde=True,stat="density")
sns.histplot(data1,ax=axs[1,1],color="red",bins=30,kde=True,stat="density")
sns.histplot(data,ax=axs[2,0],color="blue",bins=30,kde=True,stat="density")
axs[0,1].set_title("uniform_model")
axs[0,1].grid(True)
axs[0,0].set_title("normal_model")
axs[0,0].grid(True)
axs[1,1].set_title("bi_model")
axs[1,1].grid(True)
axs[2,0].set_title("multi_model")
axs[2,0].grid(True)
```
